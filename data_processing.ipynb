```python
%matplotlib inline
import pandas as pd
from pathlib import Path
from matplotlib import pyplot as plt

data = pd.read_csv(Path(r'D:\Abdul Sennain\thermal_testing_data\prelim_test.csv'), skiprows=18)
# display(data)


response = data['CH108'].dropna().multiply(-1000)
load = data['CH110'].dropna().multiply(-25000)
temp = data['CH112'].dropna()
time = data['Relative Time'].iloc[:response.size]
# display(response, load, temp, time)


fig, ax1 = plt.subplots()
ax1.plot(time/60, response, c='blue', alpha=0.8, label = 'Response')
ax1.set_xlabel('Time (min)')
ax1.set_ylabel('Response (mV)')
ax1.set_title('Response and Temperature vs. Time \nP4 ATBC')

ax2 = ax1.twinx()
ax2.plot(time/60, temp, c='red', alpha=0.8, label='Temperature')
ax2.set_ylabel('Temperature (Â°C)')
lgnd = fig.legend(loc = (.67, .775))



```
